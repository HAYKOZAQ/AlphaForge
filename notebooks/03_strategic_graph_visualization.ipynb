# Strategic Graph Visualization
This notebook visualizes the Multi-Layer Strategic Graph (Semantic 10-K, Capital 13F, Human Insiders) using NetworkX and PyVis for interactive HTML plotting.

```python
import pandas as pd
import networkx as nx
from pyvis.network import Network
from pathlib import Path
import IPython
import ast
```

### 1. Load the Graph Nodes & Edges

```python
repo_root = Path.cwd().parent
graph_dir = repo_root / "data" / "graph_models"
nodes_path = graph_dir / "graph_nodes.csv"
edges_path = graph_dir / "graph_edges.csv"

try:
    df_nodes = pd.read_csv(nodes_path)
    df_edges = pd.read_csv(edges_path)
    print(f"Loaded {len(df_nodes)} nodes and {len(df_edges)} edges.")
except Exception as e:
    print(f"Graph data not found. Please run the firm-level pipeline first. Error: {e}")
```

### 2. Build the NetworkX Graph

```python
G = nx.Graph()

for _, row in df_nodes.iterrows():
    G.add_node(row['id'], title=row['id'])

for _, row in df_edges.iterrows():
    G.add_edge(row['source'], row['target'], weight=row['weight'])

print(f"NetworkX Graph built with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")
```

### 3. Generate Interactive PyVis HTML

```python
net = Network(notebook=True, height="800px", width="100%", bgcolor="#222222", font_color="white")
net.from_nx(G)

# Optional: Add physics for better layout
net.force_atlas_2based()

html_file = "strategic_graph.html"
net.show(html_file)

IPython.display.HTML(filename=html_file)
```
